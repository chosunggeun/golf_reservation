```python
from selenium import webdriver
import time
from selenium.webdriver.common.alert import Alert
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import csv
import pandas as pd

option = webdriver.ChromeOptions()
option.add_experimental_option("excludeSwitches", ["enable-logging"])
# option.add_argument("headless")

# csv 파일 생성
final_result = []
time.sleep(1)


browser = webdriver.Chrome("./chromedriver", options=option)
browser.get(
    "https://map.naver.com/v5/?c=14188320.3206238,4344490.5447925,15,0,0,0,dh")

time.sleep(2)
place = browser.find_element(By.CSS_SELECTOR, "div.input_box>input")
place.send_keys("CC"+Keys.RETURN)
time.sleep(2)

try:
    browser.switch_to.frame("searchIframe")

    time.sleep(2)
    scroll_div = browser.find_element(By.CSS_SELECTOR,
                                      "div._1Az1K")
    # 검색 결과로 나타나는 scroll-bar 포함한 div 잡고
    # 여기서 범위 지정
    for i in range(0, 6):
        browser.execute_script("arguments[0].scrollBy(0,2000)", scroll_div)
        time.sleep(2)
        browser.execute_script("arguments[0].scrollBy(0,2000);", scroll_div)
        time.sleep(2)
        browser.execute_script("arguments[0].scrollBy(0,2000);", scroll_div)
        time.sleep(2)
        browser.execute_script("arguments[0].scrollBy(0,2000);", scroll_div)
        time.sleep(2)
        browser.execute_script("arguments[0].scrollBy(0,2000);", scroll_div)
        time.sleep(2)

        divList = browser.find_elements(
            By.CSS_SELECTOR, "li._22p-O._2NEjP")
        for i, v in enumerate(divList):
            try:
                a = v.find_elements(By.CSS_SELECTOR, "span._1U--J>a.wJICq")
                for ai in a:
                    if "예약" == ai.text:
                        name = v.find_element(
                            By.CSS_SELECTOR, "span._3Apve").text
                        print(name)
                        address = v.find_element(
                            By.CSS_SELECTOR, "span._3hCbH").text
                        print(address)
                        store_info = {
                            'place': name,
                            'address': address
                        }
                        final_result.append(store_info)
            except:
                print("패스")

        a = browser.find_elements(By.CSS_SELECTOR, "svg._3Dl4U")
        a[1].click()
        time.sleep(2)
except:
    pass

file = open('reservations.csv', mode='w', newline='')
writer = csv.writer(file)
writer.writerow(["place", "address"])
print(final_result)
for result in final_result:
    row = []
    row.append(result['place'])
    row.append(result['address'])
    writer.writerow(row)
file.close()
```
